```python
import pyzx as zx
from pyzx.circuit import Circuit

c = Circuit.from_qasm("""
OPENQASM 3;
include "stdgates.inc";
qubit[6] q;
cx q[0], q[1];
ch q[1], q[2];
cx q[2], q[3];
cx q[0], q[1];
h q[0];
cx q[0], q[4];
// h q[0];
// cx q[0], q[1];
cx q[2], q[3];
h q[2];
cx q[2], q[5];
// h q[2];
// cx q[2], q[3];
""")
```

```python
g = c.to_graph()
zx.draw(g)
```

```python
from pyzx.simplify import full_reduce

full_reduce(g)
zx.draw(g)
```
